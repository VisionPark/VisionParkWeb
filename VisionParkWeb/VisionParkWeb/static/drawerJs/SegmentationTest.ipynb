```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline 
# if you are running this code in Jupyter notebook

# reads image 'opencv-logo.png' as grayscale
img = cv2.imread('parking_sample.jpg') 

h, w, c = img.shape
print('width:  ', w)
print('height: ', h)
print('channel:', c)

# UL DL DR UR
# v1 = [837.8974963953335 , 144.4343891402715]
# v2 = [839.9108664307249 , 180.6334841628959]
# v3 = [876.8226504128982 , 181.3574660633484]
# v4 = [875.4804037226372 , 143.710407239819]

v1 =[836.7637540453075 , 181.3574660633484]
v2 = [835.1856988750193 , 142.26244343891403]
v3 = [872.2699953767915 , 141.53846153846155]
v4 = [874.6370781322237 , 181.3574660633484]

v1 = (round(v1[0]), round(v1[1]))
v2 = (round(v2[0]), round(v2[1]))
v3 = (round(v3[0]), round(v3[1]))
v4 = (round(v4[0]), round(v4[1]))


pts = np.array([v1,v2,v3,v4], np.int32)
# print(pts.shape)
# pts = pts.reshape((-1,1,2))
# print(pts.shape)

cv2.polylines(img,[pts],True,(0,255,255))

img = cv2.circle(img, v1, radius=0, color=(255, 0, 0), thickness=5)
img = cv2.circle(img, v2, radius=0, color=(255, 0, 0), thickness=5)
img = cv2.circle(img, v3, radius=0, color=(255, 0, 0), thickness=5)
img = cv2.circle(img, v4, radius=0, color=(255, 0, 0), thickness=5)


plt.figure(figsize=(25,15))
plt.imshow(img, aspect='auto')
```
